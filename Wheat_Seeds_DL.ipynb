```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset : Wheat Seeds
link : http://archive.ics.uci.edu/ml/datasets/seeds
        
To construct the data, seven geometric parameters of wheat kernels were measured:
1. area A,
2. perimeter P,
3. compactness C = 4*pi*A/P^2,
4. length of kernel,
5. width of kernel,
6. asymmetry coefficient
7. length of kernel groove.
All of these parameters are real-valued continuous.

```python
wheat_seeds=(pd.read_csv('data/seeds_data.csv',names=['area','perimeter','compactness','length_kernel','width_kernel','asymmetry_coef','length_kernel_groove','target']))
```

```python
wheat_seeds.shape
```

```python
wheat_seeds.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=wheat_seeds.drop(['target'],axis=1)
y=wheat_seeds['target']
```

```python
sns.pairplot(wheat_seeds,hue='target',size=3)
plt.plot()
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
logistic_model=LogisticRegression(multi_class='ovr')
```

```python
logistic_model.fit(x_train,y_train)
```

```python
y_pred=logistic_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
